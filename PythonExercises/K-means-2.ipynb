 Generate 200 random values from a bimodal distribution (i.e., a distribution with two peaks – in other
words, we now assume the data come from two classes that are not exactly overlapping). The simplest
way to do this is simply to generate N (with N between 1 and 200) data points from a first unimodal
distribution and then generate (200-N) data points from another unimodal distribution (with different
mean and standard deviations from the first – how different is for you to decide). Again, apply k-means
with k taking values from 5 to 1. Plot the final clustering to see what happened. Which value of k is
best? Does that make sense?

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np

#for KMeans
from sklearn.cluster import KMeans
import numpy as np
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=100, centers=1,
                       cluster_std=5, random_state=0,center_box=(1,100))

X2, y_true2 = make_blobs(n_samples=100, centers=1,
                       cluster_std=2, random_state=0,center_box=(1,100))
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50,cmap='viridis',c='black');
plt.scatter(X2[:, 0], X2[:, 1], s=50,cmap='viridis',c='red');
```

```python
A=np.concatenate((X, X2), axis=0)
```

```python
plt.scatter(A[:, 0], A[:, 1], s=50,cmap='viridis',c='black');
plt.show()
```

```python
kmeans = KMeans(n_clusters=1)
kmeans.fit(A)
y_kmeans = kmeans.predict(A)
```

```python
plt.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(A)
y_kmeans = kmeans.predict(A)

plt.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(A)
y_kmeans = kmeans.predict(A)

plt.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(A)
y_kmeans = kmeans.predict(A)

plt.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(A)
y_kmeans = kmeans.predict(A)

plt.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=50, cmap='viridis')
```
