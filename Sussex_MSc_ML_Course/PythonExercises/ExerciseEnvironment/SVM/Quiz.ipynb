```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt 




w1=7.37 
w2=1.87
def f(x1,x2):
    return w1*x1 +w2*x2
    
D =np.array([[1.2,-5.5,1],[-11.5,2.5,-1],[0.1,3.1,-1],[8.9,-0.3,1]]) 
```

```python
total_sum=0;
for i in range(len(D[:,1])):
    a = max(0, 1 - (D[i,2] * f(D[i,0],D[i,1])))
    total_sum= total_sum + a
total_sum
```
