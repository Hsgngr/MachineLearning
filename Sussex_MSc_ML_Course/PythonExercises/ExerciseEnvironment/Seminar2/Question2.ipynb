```python
import scipy.io
mat = scipy.io.loadmat('sem2_q2_data.mat')
data= mat['data'];

import matplotlib.pyplot as plt

from sklearn import preprocessing
```

```python
fig, ax = plt.subplots()
ax.scatter(data[:,0],data[:,1])
plt.xlabel('feature 1', color ='green')
plt.ylabel('feature 2', color='green')
plt.xlim(5,15)
ax.tick_params(labelcolor = 'white')
```

```python
normalized_data = preprocessing.normalize(data)
standardized_data = preprocessing.scale(data)
fig, ax = plt.subplots()
ax.scatter(data[:,0],data[:,1])
#ax.scatter(normalized_data[:,0],normalized_data[:,1],color='red')
ax.scatter(standardized_data[:,0],standardized_data[:,1],color='yellow')
plt
plt.xlabel('feature 1', color ='green')
plt.ylabel('feature 2', color='green')
ax.tick_params(labelcolor = 'white')
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(data)
X=data

plt.scatter(X[:, 0], X[:, 1], alpha=0.5)

```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)


```

```python

plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

```python
pca = PCA(n_components=2,whiten=True)
pca.fit(X)
X_whiten = pca.transform(X)
#lt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.scatter(X_whiten[:, 0], X_whiten[:, 1], alpha=0.5)
```
